## Dealing with multiple populations

The problem I'm facing now is that there exist 2 populations and I'm treating them as one. There are the recreational players and the regulars. The two most commonly used statistics in poker are called VP\$IP (Voluntarily Put Money In Pot) and PFR (Preflop Raise) should act as a good divider. I strongly suspect that VP\$IP is not normally distributed among the players because of these 2 populations (PFR might be). Hopefully this will show that there exist 2 populations. I will either attempt to label them based on my crude p-value metric from the previous notebook, or k-means clustering.

Goal: I'm not interested in recreational players so I want to make a list of players whom I want to remove.

```python
def login(number):
	db_name = 'EDAV_'+str(number) # the databases are EDAV_1,...,EDAV_4
	connection = psycopg2.connect("dbname=" + db_name + " user=postgres password=postgrespass")
	mark = connection.cursor()
	return mark 
```

```python
def rename_headers(table_headers, record): # RENAME COLUMNS
	column_names = 'select column_name from information_schema.columns where table_name = ' + table_headers
	mark.execute(column_names)
	headers = mark.fetchall()

	i=0
	for col in headers:
		col = str(col)[2:-3]
		record.rename(columns={i:col}, inplace=True)
		i += 1

def rename_custom(custom_headers, df): # RENAME COLUMNS

	i=0
	for col in custom_headers:
		df.rename(columns={i:col}, inplace=True)
		i += 1
```

For VP\$IP, the calculation is (cnt_vpip / (cnt_hands - cnt_walks)) * 100:

For PFR, the calculation is (cnt_pfr / cnt_pfr_opp) * 100

I need the following columns, which all live in the cash_cache table.

* cnt_vpip 
* cnt_hands  
* cnt_walks 
* cnt_pfr 
* cnt_pfr_opp

```python
def vpip_pfr(number):

    mark = login(number)
    distinct = ''
    column = 'id_player, id_limit, cnt_vpip, cnt_hands, cnt_walks, cnt_pfr, cnt_pfr_opp, amt_expected_bb_won'
    table = 'cash_cache'
    
    custom_headers = ('PLAYER_ID', 'LIMIT', 'cnt_vpip', 'cnt_hands', 'cnt_walks', 'cnt_pfr', 'cnt_pfr_opp', 'amt_expected_bb_won')
    
    statement = 'SELECT ' + column + ' FROM ' + table
    
    mark.execute(statement)
    
    vpip = mark.fetchall()
    vpip = pd.DataFrame(vpip)
    rename_custom(custom_headers, vpip)
    
    ## LIMIT IDs
    statement = 'SELECT id_limit,amt_bb FROM cash_limit'
    
    mark.execute(statement)
    limits = mark.fetchall()
    limits = pd.DataFrame(limits)
    
    custom_headers = ('LIMIT','BB_SIZE')
    rename_custom(custom_headers, limits)
    
    limits.to_csv('limits.csv')
    
    ## PLAYER NAMES
    statement = 'SELECT id_player, player_name FROM player'
    mark.execute(statement)
    screennames = mark.fetchall()
    screennames = pd.DataFrame(screennames)
    
    custom_headers = ('PLAYER_ID', 'NAME')
    rename_custom(custom_headers, screennames)
    
    for i in range(len(screennames.NAME)):
        name = screennames.NAME.ix[i]
        name = unicode(name, errors = 'ignore')
        screennames.NAME.ix[i] = name
    
    screennames.to_csv('screennames.csv')
    
    
    screennames = screennames.set_index('PLAYER_ID')
    limits = limits.set_index('LIMIT')
    
    limits = limits[np.logical_or(limits.BB_SIZE == 1,limits.BB_SIZE == 2)]
    
    vpip = vpip.join(limits, on='LIMIT', how = 'left')
    del vpip['LIMIT']
    
    vpip = vpip[np.logical_or(vpip.BB_SIZE == 1, vpip.BB_SIZE == 2)]
    
    vpip = vpip.join(screennames, on = 'PLAYER_ID', how = 'left')
    vpip = vpip.set_index('NAME')
    del vpip['PLAYER_ID']
    del vpip['BB_SIZE']
    vpip.to_csv('5_vpip_pfr_' + str(number) + '.csv')

for i in range(1,5):
    vpip_pfr(i)
```

We can bucket the players by the number of hands we have. 0-50, 51-250, 251-1000, 1001-5000, 5001-10000, 10001-25000, 25001,100000, 100001+

```python
def buckets(x):
    if x < 51:
        return '0-50'
    elif x < 251:
        return '51-250'
    elif x < 1001:
        return '251-1000'
    elif x < 5001:
        return '1001-5000'
    elif x < 10001:
        return '5001-10,000'
    elif x < 25001:
        return '10,001-25,000'
    elif x < 100001:
        return '25,001-100,000'
    else:
        return '100k+'

vpip_all['HAND_GROUP'] = vpip_all.cnt_hands.apply(buckets)

vpip_all.head()
```

```python
from sklearn.cluster import KMeans

def cluster_plot(scatters, clusters):
    np.random.seed(5)
    
    centers = [[1, 1], [-1, -1], [1, -1]]
    
    X = scatters[['VPIP','PFR']]
    
    est = KMeans(n_clusters=clusters)
    est.fit(X)
    labels = est.labels_
    
    labels = pd.Series(labels, index = X.index, name = 'LABEL')
    
    X = X.join(labels, how = 'left')
    X = X.join(scatters.EV_TOTAL, how = 'left')
    X = X.join(scatters.cnt_hands, how = 'left')
    

    ax = plt.subplot(3,2,clusters-1)
    fig.subplots_adjust(bottom=0.2)

    colors = brewer2mpl.get_map('Set1', 'qualitative', 8).mpl_colors
    
    for i in range(clusters):
        scatter(X[X.LABEL == i].VPIP, X[X.LABEL == i].PFR, color = colors[i], alpha = 0.5, label = 'EV/250: ' + str('%.2f' % (X[X.LABEL == i].EV_TOTAL.sum() / X[X.LABEL == i].cnt_hands.sum())))
    
    xlabel('VP$IP', fontsize = 10, fontweight = 'bold', alpha = 0.5)
    ylabel('PFR', fontsize = 10, fontweight = 'bold', alpha = 0.5)
    ax.set_title(str(clusters) + ' Clusters', fontsize = 10, fontweight = 'bold')
    ax.legend(loc = 'best')
              
    for label in ax.get_xticklines():
        label.set_visible(False)
    for label in ax.get_yticklines():
        label.set_visible(False)
    
    rstyle(ax)

fig = plt.figure(figsize = (20,16))    
for i in range(2,8):
    cluster_plot(scatters, i)
plt.suptitle('VP$IP vs. PFR For Varying Cluster Sizes \n Legend Shows Weighted Average Winrate', fontsize = 14, fontweight = 'bold')
plt.show()
```

First, VP$IP vs. winrate

```python
def plot_vpip_winrate(scatters, n, hands):
    ax = plt.subplot(3,2,n)
    
    
    xmin = scatters.VPIP.min()
    xmax = scatters.VPIP.max()
    ymin = scatters.EV_250.min()
    ymax = scatters.EV_250.max()
    
    x = np.arange(xmin,xmax,.1)
    
    colors = brewer2mpl.get_map('BrBG', 'diverging', 8).mpl_colors
    
    scatter(scatters.VPIP,scatters.EV_250, color = colors[5], alpha = 0.5)
    
    b2,b1,b0 = polyfit(scatters.VPIP, scatters.EV_250, 2) 
    plot(x, b2*x**2+b1*x+b0, color = colors[7], label = 'WAR', linewidth = 3)
    
    xlabel('VP$IP', fontsize = 10, fontweight = 'bold', alpha = 0.5)
    ylabel('BB/250', fontsize = 10, fontweight = 'bold', alpha = 0.5)
    ax.set_title('Minimum ' + str(hands) + ' hands', fontsize = 10, fontweight = 'bold')
    ax.legend(loc = 'best')
              
    for label in ax.get_xticklines():
        label.set_visible(False)
    for label in ax.get_yticklines():
        label.set_visible(False)
    
    xlim([xmin,xmax])
    ylim([ymin,ymax])
    rstyle(ax)

fig = plt.figure(figsize = (20,15))
num = 1
for i in (50, 100, 1000, 5000, 25000, 100000):
    plot_vpip_winrate(scatters[scatters.cnt_hands > i], num, i)
    num += 1
plt.suptitle('VP$IP vs. Winrate For Different Minimum Hand Cutoffs', fontsize = 14, fontweight = 'bold')
```

These match intuition quite well. Firstly, y-axis scale tightens as we get more and more samples. Most of the huge winrates we see in the smaller samples are unsustainable.

Additionally, we can see that there is typically a steep drop off after 30 VP\$IP. Most regular players do not have a VP\$IP over 30, as we can see in the minimum 100k hands graph. The highest value looks to be under 29.

There is not a particularly strong trend on the low end, but we can see the curves start lower on the left end, and peak around 25 in most cases.

The interesting exception is in the minimum 100k hand plot. We can see that for these players, there is an upward trend throughout. There are few data points and so we'll need to dig into this more later.
